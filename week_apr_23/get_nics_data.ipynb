```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.fbi.gov/file-repository/nics_firearm_checks_-_month_year.pdf/view")
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
all_a_tags = list(soup.find_all("a"))
```

```python
[line for line in all_a_tags if "nics" in str(line).lower()]
```

```python
str([line for line in all_a_tags if "nics" in str(line).lower()][0])
```

```python
bs_obj_link = [line for line in all_a_tags if ".PDF" in str(line).upper()][0]
```

```python
data_url = bs_obj_link["href"]
```

```python
data_url
```

```python
response_pdf = requests.get(data_url)
```

```python
response_pdf
```

```python
with open('nics_data.pdf', 'wb') as ofile:
    ofile.write(response_pdf.content)
```
